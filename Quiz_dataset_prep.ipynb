```python

def choose():

  import random

  # Generate a random integer, either 0 or 1
  random_bit = random.randint(0, 1)
  # if the number generated is closer to 0, choose zero. else choose 1
  if(random_bit > (1-random_bit)):
    random_bit = 1
  else:
    random_bit = 0
  return random_bit
```

```python
def choose_4():

  import random
  import math
  # Generate a random integer, either 0, 0.25, 0.5 or 1
  random_bit = random.randint(0, 1)
  v = [random_bit, abs(0.33- random_bit) ,abs(0.66- random_bit), (1- random_bit) ]
  x = v.index(min(v))
  return x
```

```python
def gen_ind():

  import random

  # Generate a random integer between 0 and 259
  ind = random.randint(0, 259)  # Use random.randint for direct range specification
  return ind

def gen_choice():
  import random
  # Generate a random integer between 0 and 259
  ch = choose_4()  # Use random.randint for direct range specification
  v = [2,3,4,5]
  return v[ch]
```

```python
def ret_int(q):
  i = 0
  while(q[i]!=None):
    if q[i].isdigit():
      if q[i-1] == "-":
        return -1*int(q[i])
      else:
        return int(q[i])
    i+=1
  return 0
```

```python
#cleaning html stuff from the prompts
import pandas as pd
from html import unescape

# Load your original CSV
df = pd.read_csv('quiz_hiprom.csv')

# Drop the 'Unnamed: 0' column if it exists
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Clean HTML artifacts in 'prompt' and 'response' columns
df['prompt'] = df['prompt'].apply(unescape)
df['response'] = df['response'].apply(unescape)

# Save the cleaned CSV
df.to_csv('quiz_hiprom1_cleaned.csv', index=True)

print("✅ Cleaned CSV saved as 'quiz_hiprom1_cleaned.csv'")
```
